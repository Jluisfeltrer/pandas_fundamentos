```python
# crear una funcion para categorizar las edades
def categorize_age(age):
    if age <= 11:
        return 'Child'
    elif age >11 and age <= 17:
        return 'Teenager'   
    elif age > 17 and age <= 59:
        return 'Adult'
    else:
        return 'Senior'
```

```python
# Aplicar la funcion a la columna 'Age2'
df_titanic0['Age_category'] = df_titanic0['Age2'].apply(categorize_age)
```

```python
#Visualizar la correlacion entre supervivencia y edad
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=df_titanic0)
plt.title('Survival by Age')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
# correlacion entre suprvvencia y sexo
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived',y='Sex_numeric', data=df_titanic0)
plt.title('Survival by SEX')
plt.xlabel('Survived')
plt.ylabel ('Sex')
plt.show() 
```

```python
#Contabilizar los datos de supervivencia
df_titanic0['Survived'].value_counts()
```

```python
sns.countplot(x='Survived', data=df_titanic0)
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
#Agregar el numero del valor a la grafica
sns.countplot(x='Survived', data=df_titanic0)
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count') 
plt.xticks(ticks=[0, 1], labels=['Sobrevivio', 'No sobrevivio'])
```

```python
# Contabilizar la supervivencia por Sex__numeric
df_titanic0.groupby('Sex')['Survived'].value_counts().unstack().plot(kind='bar', stacked=False)
plt.title('Survival Count')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Sobrevivio', 'No sobrevivio'])
plt.show()
```
